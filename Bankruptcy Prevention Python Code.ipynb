# MODEL BUILDING AND EVALUATION

```python
# Defining Features and Target
X_interaction = df_interaction.drop(columns = ["class"])
y_interaction = df_interaction["class"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Train-Test Split (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_interaction, y_interaction, test_size = 0.2, random_state = 42, stratify = y_interaction)
```

```python
from sklearn.linear_model import LogisticRegression

# Train a Logistic Regression model
logreg_model = LogisticRegression(max_iter=1000, random_state=42)
logreg_model.fit(X_train, y_train)

# Predictions
y_pred_logreg = logreg_model.predict(X_test)

# Evaluate performance
accuracy_logreg = accuracy_score(y_test, y_pred_logreg)
report_logreg = classification_report(y_test, y_pred_logreg)

print("accuracy_logreg:", accuracy_logreg)
print("report_logreg:", report_logreg)
```

```python
# Perform 5-fold cross-validation on the Logistic Regression model
from sklearn.model_selection import cross_val_score

cv_scores_logreg = cross_val_score(logreg_model, X_interaction, y_interaction, cv=5, scoring="accuracy")

# Calculate mean and standard deviation of cross-validation scores
cv_mean_logreg = cv_scores_logreg.mean()
cv_std_logreg = cv_scores_logreg.std()

print("cv_mean_logreg:", cv_mean_logreg)
print("cv_std_logreg:", cv_std_logreg)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest classifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Evaluate performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("accuracy:", accuracy)
print("report:", report)
```

```python
from sklearn.model_selection import cross_val_score

# Perform 5-fold cross-validation on the Random Forest model
cv_scores = cross_val_score(model, X_interaction, y_interaction, cv=5, scoring="accuracy")

# Calculate mean and standard deviation of cross-validation scores
cv_mean = cv_scores.mean()
cv_std = cv_scores.std()

print("cv_mean:", cv_mean)
print("cv_std:", cv_std)
```

```python
# Train a Random Forest classifier again for direct comparison
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred_rf = rf_model.predict(X_test)

# Evaluate performance
accuracy_rf = accuracy_score(y_test, y_pred_rf)
report_rf = classification_report(y_test, y_pred_rf)

# Perform cross-validation
cv_scores_rf = cross_val_score(rf_model, X_interaction, y_interaction, cv=5, scoring="accuracy")
cv_mean_rf = cv_scores_rf.mean()
cv_std_rf = cv_scores_rf.std()

# Store results for comparison
print("Random Forest Test Accuracy:", accuracy_rf)
print("Random Forest CV Mean Accuracy:", cv_mean_rf)
print("Random Forest CV Std Dev:", cv_std_rf)
print("Classification Report:", report_rf)
```

### Model Interpretation

```python
# ---- 1. Logistic Regression Coefficients ---- #
coef_values = logreg_model.coef_[0]
features = X_interaction.columns

# Create a dataframe for visualization
coef_df = pd.DataFrame({'Feature': features, 'Coefficient': coef_values})
coef_df = coef_df.sort_values(by='Coefficient', ascending=False)

# Plot Coefficients
plt.figure(figsize=(8, 5))
sns.barplot(x='Coefficient', y='Feature', data=coef_df, palette='coolwarm')
plt.axvline(0, color='black', linewidth=1)  # Add a vertical line at 0 for reference
plt.title("Logistic Regression Coefficients (Feature Impact)")
plt.xlabel("Impact on Bankruptcy Probability")
plt.ylabel("Feature")
plt.show()
```

```python
# ---- 2. Random Forest Feature Importance ---- #
rf_importance = rf_model.feature_importances_
importance_df_rf = pd.DataFrame({'Feature': X_interaction.columns, 'Importance': rf_importance})
importance_df_rf = importance_df_rf.sort_values(by='Importance', ascending=False)

# Plot Feature Importance
plt.figure(figsize=(8, 5))
sns.barplot(x='Importance', y='Feature', data=importance_df_rf, palette='viridis')
plt.title("Random Forest Feature Importance")
plt.xlabel("Importance Score")
plt.ylabel("Feature")
plt.show()
```

```python
# ---- 3. Confusion Matrix ---- #
from sklearn.metrics import confusion_matrix

y_pred_log = logreg_model.predict(X_test)
conf_matrix = confusion_matrix(y_test, y_pred_log)

# Plot Confusion Matrix
plt.figure(figsize=(5, 4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Non-Bankrupt', 'Bankrupt'], yticklabels=['Non-Bankrupt', 'Bankrupt'])
plt.title("Confusion Matrix for Logistic Regression")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

### AUC-ROC Curve

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Predict probabilities for Logistic Regression
y_pred_proba_logreg = logreg_model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC for Logistic Regression
fpr_logreg, tpr_logreg, _ = roc_curve(y_test, y_pred_proba_logreg)
auc_logreg = roc_auc_score(y_test, y_pred_proba_logreg)

plt.figure(figsize=(8, 6))
plt.plot(fpr_logreg, tpr_logreg, label=f'Logistic Regression (AUC = {auc_logreg:.2f})', color='blue')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')  # Diagonal line for random guessing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()

print(f'Logistic Regression AUC Score: {auc_logreg:.2f}')

# If Random Forest is implemented, compute AUC for it as well
if 'rf_model' in locals():
    y_pred_proba_rf = rf_model.predict_proba(X_test)[:, 1]
    fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_proba_rf)
    auc_rf = roc_auc_score(y_test, y_pred_proba_rf)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr_rf, tpr_rf, label=f'Random Forest (AUC = {auc_rf:.2f})', color='green')
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve - Random Forest')
    plt.legend()
    plt.show()

    print(f'Random Forest AUC Score: {auc_rf:.2f}')
```
